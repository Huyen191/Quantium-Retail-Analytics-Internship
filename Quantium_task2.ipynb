```python
# Nhom du lieu theo cua hang va thang
monthly_metrics = df.groupby(['STORE_NBR', 'MONTH']).agg(
    Total_Sales=('TOT_SALES', 'sum'),           # Tổng doanh thu trong tháng
    Total_Customers=('LYLTY_CARD_NBR', 'nunique'), # Số khách hàng duy nhất trong tháng
    Total_Transactions=('TXN_ID', 'count')      # Tổng số giao dịch trong tháng
).reset_index()
```

```python
#Tinh trung binh giao dich cua moi khach hang
monthly_metrics['Avg_Transaction_Per_Cust'] = monthly_metrics['Total_Transactions'] / monthly_metrics['Total_Customers']
```

```python
def find_best_control_store(trial_store, all_stores, monthly_metrics):
    """
    Tìm cửa hàng đối chứng phù hợp nhất cho trial_store.
    
    Parameters:
    - trial_store: Mã cửa hàng thử nghiệm (ví dụ: 77)
    - all_stores: Danh sách tất cả cửa hàng
    - monthly_metrics: DataFrame đã tổng hợp theo tháng (có cột Avg_Transaction_Per_Cust)
    
    Returns:
    - Mã cửa hàng đối chứng tốt nhất và điểm correlation
    """
    # Loại bỏ các cửa hàng thử nghiệm khỏi danh sách ứng viên
    trial_stores = [77, 86, 88]
    candidate_stores = [s for s in all_stores if s not in trial_stores]
    
    # Lấy dữ liệu của cửa hàng thử nghiệm
    trial_data = monthly_metrics[monthly_metrics['STORE_NBR'] == trial_store].set_index('MONTH')
    
    best_store = None
    best_score = -1  # Correlation cao nhất sẽ được chọn
    
    # Duyệt qua từng cửa hàng ứng viên
    for store in candidate_stores:
        # Lấy dữ liệu của cửa hàng ứng viên
        control_data = monthly_metrics[monthly_metrics['STORE_NBR'] == store].set_index('MONTH')
        
        # Tìm các tháng chung giữa trial và control
        common_months = trial_data.index.intersection(control_data.index)
        
        # Bỏ qua nếu không có đủ tháng chung (ít nhất 3 tháng)
        if len(common_months) < 3:
            continue
        
        # Lấy dữ liệu chung (chỉ những tháng có trong cả trial và control)
        trial_subset = trial_data.loc[common_months]
        control_subset = control_data.loc[common_months]
        
        # Tính correlation cho từng chỉ số
        sales_corr, _ = pearsonr(trial_subset['Total_Sales'], control_subset['Total_Sales'])
        cust_corr, _ = pearsonr(trial_subset['Total_Customers'], control_subset['Total_Customers'])
        trans_corr, _ = pearsonr(trial_subset['Avg_Transaction_Per_Cust'], 
                                 control_subset['Avg_Transaction_Per_Cust'])
        
        # Tính điểm trung bình (càng cao càng tốt)
        avg_corr = (sales_corr + cust_corr + trans_corr) / 3
        
        # Cập nhật cửa hàng tốt nhất
        if avg_corr > best_score:
            best_score = avg_corr
            best_store = store
    
    return best_store, best_score
```

```python
# Danh sách tất cả cửa hàng
all_stores = monthly_metrics['STORE_NBR'].unique()

# Danh sách cửa hàng thử nghiệm
trial_stores = [77, 86, 88]

# Tìm cửa hàng đối chứng cho từng cửa hàng thử nghiệm
control_mapping = {}
print("Cửa hàng đối chứng được chọn:")
for trial in trial_stores:
    control, score = find_best_control_store(trial, all_stores, monthly_metrics)
    control_mapping[trial] = control
    print(f"- Cửa hàng thử nghiệm {trial} → Cửa hàng đối chứng {control} (Correlation: {score:.3f})")
```

```python
# Thời gian thử nghiệm (giả sử từ tháng 6 đến tháng 8 năm 2023)
trial_months = pd.period_range(start='2019-03', end='2019-05', freq='M')

print("\nSo sánh hiệu quả trong thời gian thử nghiệm:")

for trial_store in trial_stores:
    control_store = control_mapping[trial_store]
    
    # Lọc dữ liệu trong thời gian thử nghiệm
    trial_data = monthly_metrics[
        (monthly_metrics['STORE_NBR'] == trial_store) &
        (monthly_metrics['MONTH'].isin(trial_months))
    ]
    
    control_data = monthly_metrics[
        (monthly_metrics['STORE_NBR'] == control_store) &
        (monthly_metrics['MONTH'].isin(trial_months))
    ]
    
    # Tính trung bình các chỉ số
    avg_sales_trial = trial_data['Total_Sales'].mean()
    avg_sales_control = control_data['Total_Sales'].mean()
    
    avg_customers_trial = trial_data['Total_Customers'].mean()
    avg_customers_control = control_data['Total_Customers'].mean()
    
    avg_trans_trial = trial_data['Avg_Transaction_Per_Cust'].mean()
    avg_trans_control = control_data['Avg_Transaction_Per_Cust'].mean()
    
    # Tính phần trăm thay đổi
    sales_change = (avg_sales_trial - avg_sales_control) / avg_sales_control * 100
    customers_change = (avg_customers_trial - avg_customers_control) / avg_customers_control * 100
    trans_change = (avg_trans_trial - avg_trans_control) / avg_trans_control * 100
    
    print(f"\nCửa hàng {trial_store} vs {control_store}:")
    print(f"  - Doanh thu trung bình: {avg_sales_trial:.2f} vs {avg_sales_control:.2f} ({sales_change:+.1f}%)")
    print(f"  - Số khách trung bình: {avg_customers_trial:.2f} vs {avg_customers_control:.2f} ({customers_change:+.1f}%)")
    print(f"  - Giao dịch/khách:    {avg_trans_trial:.2f} vs {avg_trans_control:.2f} ({trans_change:+.1f}%)")
```

```python
#so sánh doanh thu theo tháng
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

for i, trial_store in enumerate([77, 86, 88]):
    control_store = control_mapping[trial_store]
    
    # Lọc dữ liệu theo tháng
    trial_data = monthly_metrics[
        monthly_metrics['STORE_NBR'] == trial_store
    ].copy()
    control_data = monthly_metrics[
        monthly_metrics['STORE_NBR'] == control_store
    ].copy()
    
    # Thêm cột để phân biệt trial và control
    trial_data['Type'] = 'Trial'
    control_data['Type'] = 'Control'
    
    # Ghép dữ liệu lại
    combined_data = pd.concat([trial_data, control_data])
    
    combined_data['MONTH'] = combined_data['MONTH'].astype(str)
    
    # Vẽ biểu đồ 
    sns.lineplot(
        data=combined_data,
        x='MONTH', 
        y='Total_Sales', 
        hue='Type',
        marker='o',
        ax=ax[i]
    )
    
    ax[i].set_title(f'Cửa hàng {trial_store} vs {control_store}')
    ax[i].set_ylabel('Doanh thu')
    ax[i].grid(True)

fig.suptitle('So sánh doanh thu theo tháng giữa cửa hàng thử nghiệm và đối chứng', 
             fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
